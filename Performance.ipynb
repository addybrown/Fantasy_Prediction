```python
Player=[]
Triple_Doubles=[]
Year=[]

#range in between data is available 
i=2019

    #clear all variables on iteration
a=[]
    
    #gets request from site to web-scrap
res=requests.get('http://www.landofbasketball.com/year_by_year_stats/'+str(i-1)+'_'+str(i)+'_triple_doubles_rs.htm')
soup=bs4.BeautifulSoup(res.text,'lxml')
storage=[]
    
    #saves xml format
for line in soup:
    storage.append(line)
        
    #takes second value of xml code as 1st line is just 'html'
a=str(storage[1]).split('style="width: 140px')
    
    #clear Player and TripleDouble Variables
    #Player=[]
    #Triple_Doubles=[]
    #Year=[]
    
    #Finds values inbetween harded coded xml elements that point to player name and triple doubles
for value in a[1:]:
    d1 = re.search('htm">(.*)</a>', value)
    d2= re.search('class="negri">(.*)</td>',value)
    Player.append(d1.group(1))
    Triple_Doubles.append(d2.group(1))
    Year.append(i)
    
d3 = {'Player': Player, 'TD':  Triple_Doubles}
Triple_Doubles=pd.DataFrame(d3)
```

```python
#Web Scrape the Current Stats from Basketball Reference Advanced Statistics 
res=requests.get('https://www.basketball-reference.com/leagues/NBA_2019_totals.html')
soup=bs4.BeautifulSoup(res.text,'lxml')
soup=str(soup)
g=soup.split('data-stat=')

Player=[]
Pos=[]
Age=[]
Team=[]
Games=[]
Minutes=[]
PTS=[]
AST=[]
BLK=[]
STL=[]
TRB=[]
TOV=[]
Three_P=[]
FT=[]
FTA=[]

#print(g)

for value in g[30:]:
    player = re.search('.html">(.*)</a></td><td class="center "', value)
    pos = re.search('"pos">(.*)</td>', value)
    age = re.search('"age">(.*)</td>', value)
    team = re.search('"/teams/(.*)/2019.html"', value)
    gms = re.search('"g">(.*)</td>', value)
    minutes = re.search('"mp">(.*)</td>', value)
    
    
    pts =  re.search('"pts">(.*)</td>', value)
    tov= re.search('"tov">(.*)</td>', value)
    ast=re.search('"ast">(.*)</td>', value)
    blk = re.search('"blk">(.*)</td>', value)
    stl = re.search('"stl">(.*)</td>', value)
    three_p=re.search('"fg3">(.*)</td>', value)
    #print(three_p)
    ft = re.search('"ft">(.*)</td>', value)
    fta=re.search('"fta">(.*)</td>', value)
    trb=re.search('"trb">(.*)</td>', value)
    
    

    if player is not None:
        Player.append(player.group(1)) 
    if pos is not None:
        Pos.append(pos.group(1))
    if age is not None:
        Age.append(age.group(1))
    if team is not None:
        Team.append(team.group(1))
    if gms is not None:
        Games.append(gms.group(1))
    if minutes is not None:
        Minutes.append(minutes.group(1))
        
        
    if pts is not None:
        PTS.append(pts.group(1))
    if ast is not None:
        AST.append(ast.group(1))
    if blk is not None:
        BLK.append(blk.group(1))
    if stl is not None:
        STL.append(stl.group(1))
    if three_p is not None:
        Three_P.append(three_p.group(1))
    if ft is not None:
        FT.append(ft.group(1))
    if fta is not None:
        FTA.append(fta.group(1))
    if tov is not None:
        TOV.append(tov.group(1))
    if trb is not None:
        TRB.append(trb.group(1))
    #print(three_p)
temp=pd.DataFrame()

len(Three_P)

temp['Player']=Player
temp['Pos']= Pos
temp['Age']= list(map(int, Age))
temp['MP']=list(map(int,Minutes))
#temp['Team']= Team
temp['Games']= list(map(int, Games))
temp['PTS']=list(map(float, PTS))
temp['AST']=list(map(float,AST))
temp['BLK']=list(map(float,BLK))
temp['STL']= list(map(int, STL))
temp['3P']=list(map(float, Three_P))
temp['FT']=list(map(float, FT))
temp['FTA']=list(map(float,  FTA))
temp['TOV']=list(map(float,TOV))
temp['TRB']=list(map(float,TRB))


temp['MPG']=temp['MP']/temp['Games']
temp['FTM']=temp['FTA']-temp['FT']
                  
regular_2019=temp[['Player','Games','PTS','AST','TRB','BLK','STL','3P','FT','FTM','TOV']]
```

```python
regular_2019=pd.merge(regular_2019,Triple_Doubles,how='left',on='Player')
regular_2019=pd.merge(regular_2019,Double_Doubles,how='left',on='Player')
regular_2019['TD'].fillna(0,inplace=True)
regular_2019['DD'].fillna(0,inplace=True)
```

```python
regular_2019['Fantasy_Pts']=regular_2019['PTS']+1.5*regular_2019['AST']+regular_2019['TRB']+regular_2019['BLK']+2.5*regular_2019['STL']\
                            -1.5*regular_2019['TOV']+0.5*regular_2019['FT']-0.75*regular_2019['FTM']+3*regular_2019['TD'].astype(float)\
                            +regular_2019['DD'].astype(float)
```

```python
regular_2019['Fantasy_PPG']=regular_2019['Fantasy_Pts']/regular_2019['Games']
```

```python
import json
from keras.models import model_from_json
```

```python
json_file=open('model.json','r')
loaded_model_json=json_file.read()
json_file.close()
loaded_model=model_from_json(loaded_model_json)

loaded_model.load_weights("model.h5")
print("Model Has Been Loaded")
```

```python
advanced_2019.index=advanced_2019['Player']
advanced_2019.drop('Player',axis=1,inplace=True)
```

```python
regular_2019['Predictions_Advanced']=loaded_model.predict(advanced_2019)
```

```python
regular_2019.sort_values('Fantasy_PPG',ascending=False).head()
final=pd.merge(regular_2019[['Player','Fantasy_PPG','Predictions_Advanced']],predictions,how='right',on='Player')
```

## Fantasy_PPG_Prediction is the pre-season points generated per game for each player

## Predictions_Advanced is predictions using the current advanced stats

## Fantasy_PPG is the actual Fantasy Points Per Game 

```python
final.sort_values(by='Fantasy_PPG',ascending=False,inplace=True)
final.index=range(0,len(final))
final.head()
```

```python
final.to_csv('data-files/Performance.csv')
```

```python
predictions_Error=np.mean(np.abs(final['Fantasy_PPG']-final['Fantasy_PPG_Prediction']))
advanced_Error=np.mean(np.abs(final['Fantasy_PPG']-final['Predictions_Advanced']))

print('The Overall MSE for the Prediction Error is: ' + str(round(predictions_Error,2)))
print('The Overall MSE for the Advanced Error is: ' + str(round(advanced_Error,2)))
```
