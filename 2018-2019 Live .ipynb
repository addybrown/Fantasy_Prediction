```python
import pandas as pd 
import numpy as np
import re 
import bs4 
import requests
```

```python
res=requests.get('https://www.basketball-reference.com/leagues/NBA_2019_advanced.html')
soup=bs4.BeautifulSoup(res.text,'lxml')
soup=str(soup)
g=soup.split('data-stat=')
```

```python
Player=[]
Pos=[]
Age=[]
Team=[]
Games=[]
Minutes=[]
PER=[]
OWS=[]
DWS=[]
WS_PER_48=[]
BPM=[]
VORP=[]
```

```python
for value in g[30:]:
    player = re.search('.html">(.*)</a></td><td class="center "', value)
    pos = re.search('"pos">(.*)</td>', value)
    age = re.search('"age">(.*)</td>', value)
    team = re.search('"/teams/(.*)/2019.html"', value)
    gms = re.search('"g">(.*)</td>', value)
    minutes = re.search('"mp">(.*)</td>', value)
    per = re.search('"per">(.*)</td>', value)
    ows = re.search('"ows">(.*)</td>', value)
    dws = re.search('"dws">(.*)</td>', value)
    ws_per_48=re.search('"ws_per_48">(.*)</td>', value)
    bpm = re.search('"bpm">(.*)</td>', value)
    vorp = re.search('"vorp">(.*)</td>', value)

    if player is not None:
        Player.append(player.group(1)) 
    if pos is not None:
        Pos.append(pos.group(1))
    if age is not None:
        Age.append(age.group(1))
    if team is not None:
        Team.append(team.group(1))
    if gms is not None:
        Games.append(gms.group(1))
    if minutes is not None:
        Minutes.append(minutes.group(1))
    if per is not None:
        PER.append(per.group(1))
    if ows is not None:
        OWS.append(ows.group(1))
    if dws is not None:
        DWS.append(dws.group(1))
    if ws_per_48 is not None:
        WS_PER_48.append(ws_per_48.group(1))
    if bpm is not None:
        BPM.append(bpm.group(1))
    if vorp is not None:
        VORP.append(vorp.group(1))
```

```python
df=pd.DataFrame()

df['Player']=Player
df['Pos']= Pos
df['Age']= list(map(int, Age))
df['Team']= Team
df['Games']= list(map(int, Games))
df['MP']= list(map(int, Minutes))
df['PER']=list(map(float, PER))
df['BPM']=list(map(float, BPM))
df['WS/48']=list(map(float,  WS_PER_48))
```

```python
writer = pd.ExcelWriter('2018_2019_Live.xlsx')
df.to_excel(writer,'Sheet1')
writer.save()
```
